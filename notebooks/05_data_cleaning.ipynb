```python
import pandas as pd
import numpy as np
from tqdm import tqdm

max_missing_percentage = 30
print("Calculating initial statistics...")
ep_df['date'] = pd.to_datetime(ep_df['timestamp']).dt.date
day_stats = ep_df.groupby(['ride_name', 'date']).agg(
    total_entries=('wait_time', 'count'),
    missing_entries=('wait_time', lambda x: x.isna().sum())
).reset_index()
day_stats['missing_percentage'] = (day_stats['missing_entries'] / day_stats['total_entries'] * 100).round(2)
day_stats['drop'] = day_stats['missing_percentage'] > max_missing_percentage
drop_summary = day_stats.groupby('ride_name').agg(
    days_to_drop=('drop', 'sum'),
    total_days=('drop', 'count')
).reset_index()
drop_summary['drop_percentage'] = (drop_summary['days_to_drop'] / drop_summary['total_days'] * 100).round(2)
drop_summary = drop_summary.sort_values('drop_percentage', ascending=False)

print(f"Summary of days to drop by ride (threshold: >{max_missing_percentage}% missing data):")
print("=============================================================")
for _, row in drop_summary.iterrows():
    print(f"{row['ride_name']:<30} | {row['days_to_drop']:<12} | {row['total_days']:<10} | {row['drop_percentage']}%")
total_days_to_drop = day_stats['drop'].sum()
total_days = len(day_stats)
drop_percentage = (total_days_to_drop / total_days * 100).round(2)
print(f"\nOverall: Would drop {total_days_to_drop} out of {total_days} ride-days ({drop_percentage}%)")
```

```python

# MEMORY-EFFICIENT IMPLEMENTATION
print("\n=== MEMORY-EFFICIENT 30-MINUTE BUCKET FILLING ===")

# Step 1: Add datetime components without creating a full copy
print("Preparing data...")
# Make sure tqdm is properly imported for multiple progress bars
try:
    from tqdm.auto import tqdm  # Use auto for better notebook integration
except ImportError:
    # If tqdm.auto is not available, fall back to regular tqdm
    from tqdm import tqdm
ep_df['datetime'] = pd.to_datetime(ep_df['timestamp'])
ep_df['date'] = ep_df['datetime'].dt.date
ep_df['time_bucket'] = ep_df['datetime'].dt.floor('30min')

# Step 2: Count original missing values
original_missing = ep_df['wait_time'].isna().sum()
print(f"Original missing values: {original_missing}")

# Step 3: Process in chunks by ride to reduce memory usage
print("Processing time buckets...")
filled_count = 0

# Get unique ride names
unique_rides = ep_df['ride_name'].unique()
total_rides = len(unique_rides)

for ride_idx, ride_name in enumerate(unique_rides):
    print(f"Processing ride {ride_idx+1}/{total_rides}: {ride_name}")
    
    # Get data for just this ride
    ride_df = ep_df[ep_df['ride_name'] == ride_name].copy()
    
    # Skip if no missing values
    missing_mask = ride_df['wait_time'].isna() & ~ride_df['closed']
    if not missing_mask.any():
        continue
    
    # Get unique dates for this ride
    unique_dates = ride_df['date'].unique()
    
    # Process each date with progress bar
    for date in tqdm(unique_dates, desc=f"Processing {ride_name}", leave=False):
        # Get data for this ride and date
        day_data = ride_df[ride_df['date'] == date]
        
        # Skip if no missing values
        if not (day_data['wait_time'].isna() & ~day_data['closed']).any():
            continue
        
        # Get unique time buckets for this day
        buckets = day_data['time_bucket'].unique()
        
        # Process each bucket
        for bucket in buckets:
            # Get data for this bucket
            bucket_data = day_data[day_data['time_bucket'] == bucket]
            
            # Skip if no missing values or all closed
            missing_in_bucket = bucket_data['wait_time'].isna() & ~bucket_data['closed']
            if not missing_in_bucket.any():
                continue
                
            # Sort by time within bucket
            bucket_data = bucket_data.sort_values('datetime')
            
            # Find valid values (not null, not closed)
            valid_mask = bucket_data['wait_time'].notna() & ~bucket_data['closed']
            if not valid_mask.any():
                continue
            
            # Process each missing value
            for idx in bucket_data[missing_in_bucket].index:
                current_time = bucket_data.loc[idx, 'datetime']
                
                # Find valid entries prior to this time
                valid_before = bucket_data[
                    valid_mask & 
                    (bucket_data['datetime'] < current_time)
                ]
                
                if not valid_before.empty:
                    # Get the most recent entry
                    most_recent = valid_before.iloc[-1]
                    # Update directly in the original dataframe
                    ep_df.loc[idx, 'wait_time'] = most_recent['wait_time']
                    filled_count += 1
    
    # Explicitly free memory
    del ride_df
    
    # Print overall progress
    print(f"Progress: {filled_count} values filled so far ({filled_count/original_missing*100:.2f}% complete)")

print(f"Filled {filled_count} values out of {original_missing} missing values")

# Step 4: Recalculate statistics
print("Calculating final statistics...")
day_stats_filled = ep_df.groupby(['ride_name', 'date']).agg(
    total_entries=('wait_time', 'count'),
    missing_entries=('wait_time', lambda x: x.isna().sum())
).reset_index()

day_stats_filled['missing_percentage'] = (day_stats_filled['missing_entries'] / day_stats_filled['total_entries'] * 100).round(2)
day_stats_filled['drop'] = day_stats_filled['missing_percentage'] > max_missing_percentage

drop_summary_filled = day_stats_filled.groupby('ride_name').agg(
    days_to_drop=('drop', 'sum'),
    total_days=('drop', 'count')
).reset_index()

drop_summary_filled['drop_percentage'] = (drop_summary_filled['days_to_drop'] / drop_summary_filled['total_days'] * 100).round(2)
drop_summary_filled = drop_summary_filled.sort_values('drop_percentage', ascending=False)

# Step 5: Print summary
print(f"\nSummary of days to drop by ride after filling with 30-min buckets (threshold: >{max_missing_percentage}% missing data):")
print("=============================================================")
for _, row in drop_summary_filled.iterrows():
    print(f"{row['ride_name']:<30} | {row['days_to_drop']:<12} | {row['total_days']:<10} | {row['drop_percentage']}%")

total_days_to_drop_filled = day_stats_filled['drop'].sum()
total_days_filled = len(day_stats_filled)
drop_percentage_filled = (total_days_to_drop_filled / total_days_filled * 100).round(2)

print(f"\nOverall: Would drop {total_days_to_drop_filled} out of {total_days_filled} ride-days ({drop_percentage_filled}%)")
print(f"Original missing values: {original_missing} ({(original_missing/len(ep_df)*100):.2f}%)")
print(f"Values filled: {filled_count} ({(filled_count/len(ep_df)*100):.2f}%)")
print(f"Remaining missing values: {original_missing - filled_count} ({((original_missing - filled_count)/len(ep_df)*100):.2f}%)")

# Calculate improvement
days_saved = total_days_to_drop - total_days_to_drop_filled
percent_reduction = (days_saved / total_days_to_drop * 100).round(2) if total_days_to_drop > 0 else 0
print(f"\nDays saved from dropping: {days_saved} ({percent_reduction}% reduction)")
```

```python
import pandas as pd
import numpy as np
from tqdm import tqdm  # Import tqdm for progress bar

def count_weather_drops(ep_df):
    """
    Analyzes how many records would be dropped due to missing weather data
    after attempting to fill from same timestamp.
    
    Args:
        ep_df: Original dataframe with theme park ride data
        
    Returns:
        Dictionary with counts and statistics
    """
    df = ep_df.copy()
    
    total_rows = len(df)
    
    print("Step 1/4: Counting rows with missing weather data...")
    
    weather_cols = ['temperature', 'rain', 'wind']
    
    missing_per_column = {col: df[col].isna().sum() for col in weather_cols}
    
    rows_with_missing_weather = df[df[weather_cols].isna().any(axis=1)]
    
    print("Step 2/4: Simulating filling weather data from same timestamp...")
    df_after_fill = df.copy()
    
    timestamp_stats = {}
    unfillable_timestamps = set()
    
    print("Step 3/4: Analyzing timestamps...")
    # For each timestamp, check if any ride has complete weather data
    for timestamp in tqdm(df['timestamp'].unique(), desc="Processing timestamps"):
        time_slice = df[df['timestamp'] == timestamp]
        

        has_complete_data = False
        for col in weather_cols:
            if not time_slice[col].isna().all():
                has_complete_data = True
                break
        
        rides_at_timestamp = len(time_slice)
        missing_at_timestamp = len(time_slice[time_slice[weather_cols].isna().any(axis=1)])
        
        timestamp_stats[timestamp] = {
            'rides': rides_at_timestamp,
            'missing_weather': missing_at_timestamp,
            'has_complete_data': has_complete_data
        }
        
        if not has_complete_data:
            unfillable_timestamps.add(timestamp)
    
    print("Step 4/4: Calculating final statistics...")

    unfillable_rows = df[df['timestamp'].isin(unfillable_timestamps)]
    
    still_missing_after_fill = df_after_fill[df_after_fill[weather_cols].isna().any(axis=1)]
    
    pct_unfillable = (len(unfillable_rows) / total_rows) * 100 if total_rows > 0 else 0
    percentage_rows_dropped = (len(still_missing_after_fill) / total_rows) * 100 if total_rows > 0 else 0
    

    results = {
        'total_rows': total_rows,
        'missing_per_column': missing_per_column,
        'rows_with_any_missing_weather': len(rows_with_missing_weather),
        'distinct_timestamps': len(timestamp_stats),
        'timestamps_with_no_weather_data': len(unfillable_timestamps),
        'rows_that_will_be_dropped': len(still_missing_after_fill),
        'unfillable_timestamps': sorted(unfillable_timestamps) if len(unfillable_timestamps) < 10 else f"{len(unfillable_timestamps)} timestamps",
        'unfillable_percentage': pct_unfillable,
        'percentage_rows_dropped': percentage_rows_dropped  # Added this field
    }
    
    return results


drop_analysis = count_weather_drops(ep_df)
print(f"Rows that will be dropped: {drop_analysis['rows_that_will_be_dropped']} ({drop_analysis['percentage_rows_dropped']:.2f}%)")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

def visualize_missing_weather_by_month(df):
    """
    Visualizes missing weather data by month since 2017.
    
    Args:
        df: DataFrame with theme park ride data including timestamp and weather columns
    """
    df_copy = df.copy()
    
    if not pd.api.types.is_datetime64_any_dtype(df_copy['timestamp']):
        df_copy['timestamp'] = pd.to_datetime(df_copy['timestamp'])
    
    df_filtered = df_copy[df_copy['timestamp'].dt.year >= 2017]
    
    df_filtered['year'] = df_filtered['timestamp'].dt.year
    df_filtered['month'] = df_filtered['timestamp'].dt.month
    
    df_filtered['year_month'] = df_filtered['timestamp'].dt.strftime('%Y-%m')
    
    weather_cols = ['temperature', 'rain', 'wind']
    
    missing_data = []
    
    for year_month in sorted(df_filtered['year_month'].unique()):
        year, month = year_month.split('-')
        monthly_data = df_filtered[df_filtered['year_month'] == year_month]
        total_records = len(monthly_data)
        
        month_missing = {
            'year': int(year),
            'month': int(month),
            'year_month': year_month,
            'total_records': total_records
        }
        

        for col in weather_cols:
            missing_count = monthly_data[col].isna().sum()
            missing_percentage = (missing_count / total_records) * 100 if total_records > 0 else 0
            month_missing[f'{col}_missing_pct'] = missing_percentage
        
        any_missing = monthly_data[monthly_data[weather_cols].isna().any(axis=1)].shape[0]
        month_missing['any_weather_missing_pct'] = (any_missing / total_records) * 100 if total_records > 0 else 0
        
        missing_data.append(month_missing)
    

    missing_df = pd.DataFrame(missing_data)
    
    plt.figure(figsize=(14, 10))
    

    plt.subplot(2, 1, 1)
    for col in weather_cols:
        plt.plot(missing_df['year_month'], missing_df[f'{col}_missing_pct'], 
                 marker='o', linestyle='-', label=f'{col}')
    
    plt.title('Percentage of Missing Weather Data by Month (Since 2017)', fontsize=14)
    plt.xlabel('Year-Month', fontsize=12)
    plt.ylabel('Missing Data (%)', fontsize=12)
    plt.xticks(rotation=90)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    

    for col in weather_cols:
        col_missing = f'{col}_missing_pct'
        peak_months = missing_df.nlargest(3, col_missing)
        for _, row in peak_months.iterrows():
            if row[col_missing] > 10: 
                plt.annotate(f"{row[col_missing]:.1f}%", 
                             xy=(row['year_month'], row[col_missing]),
                             xytext=(0, 10), textcoords='offset points',
                             ha='center', fontsize=8)
    

    plt.subplot(2, 1, 2)
    
    heatmap_data = missing_df.pivot_table(
        index='month', 
        columns='year',
        values='any_weather_missing_pct',
        aggfunc='mean'
    )
    
    sns.heatmap(heatmap_data, cmap='YlOrRd', annot=True, fmt='.1f', linewidths=.5)
    plt.title('Percentage of Records with Any Missing Weather Data by Month and Year', fontsize=14)
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Month', fontsize=12)
    
    month_names = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
                  7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}
    plt.yticks(np.arange(0.5, 12.5), [month_names[i] for i in range(1, 13)])
    
    plt.tight_layout()
    plt.savefig('missing_weather_visualization.png', dpi=300, bbox_inches='tight')
    plt.show()
    

    return missing_df


def print_worst_months(missing_df, top_n=5):
    """Prints the months with the highest percentage of missing weather data"""
    print(f"\nTop {top_n} Months with Highest Missing Weather Data:")
    

    worst_months = missing_df.nlargest(top_n, 'any_weather_missing_pct')
    print("\nMonths with highest percentage of ANY missing weather data:")
    for _, row in worst_months.iterrows():
        print(f"{row['year_month']}: {row['any_weather_missing_pct']:.2f}% ({row['total_records']} records)")
    

    for col in ['temperature', 'rain', 'wind']:
        col_missing = f'{col}_missing_pct'
        worst_months = missing_df.nlargest(top_n, col_missing)
        print(f"\nMonths with highest percentage of missing {col} data:")
        for _, row in worst_months.iterrows():
            print(f"{row['year_month']}: {row[col_missing]:.2f}% ({row['total_records']} records)")

missing_analysis = visualize_missing_weather_by_month(ep_df)
print_worst_months(missing_analysis)
```
